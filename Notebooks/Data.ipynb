```python
def plot_all_classes():
    imarray=[]
    label_array=['0:T-shirt','1:Trouser','2:Pullover','3:Dress','4:Coat',
                  '5:Sandal','6:Shirt','7:Sneaker','8:Bag','9:Boot']

    i=0
    j=0
    class_left=list(range(0,10))
    while len(class_left):
        if y_train[i] ==j :
            class_left.remove(j)
            
            plt.imshow(x_train[i],cmap='gray')
            plt.title('Class'+str(label_array[j]))
            plt.show()
            imarray.append(np.expand_dims(x_train[i],axis=-1))
            j+=1
            
        i+=1 
        
    return imarray, label_array

        
    
```

```python
imba,lba=plot_all_classes()
```
