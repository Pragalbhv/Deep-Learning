```python
def plot_all_classes():
    i=0
    class_left=list(range(1,10))
    while len(class_left):
        if y_train[i] in class_left:
            class_left.remove(y_train[i])
            plt.imshow(x_train[i],cmap='gray')
            plt.title('Class'+str(y_train[i]))
            plt.show()
        i+=1   

        
    
```

```python
plot_all_classes()
```
